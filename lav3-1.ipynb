```python
# -*- coding: utf-8 -*-
"""Лабораторная работа по классификации данных"""

# Импорт необходимых библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Для классических методов
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                           f1_score, roc_auc_score, classification_report, 
                           confusion_matrix, roc_curve)

# Наивные Байесовские классификаторы
from sklearn.naive_bayes import GaussianNB, MultinomialNB, ComplementNB, BernoulliNB

# Другие классификаторы
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Для нейронной сети
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import TensorBoard, EarlyStopping
import datetime

# ============================================================================
# 1. ВЫБОР И ПОДГОТОВКА ДАТАСЕТА
# ============================================================================

print("=" * 80)
print("1. ВЫБОР И ПОДГОТОВКА ДАТАСЕТА")
print("=" * 80)

# Используем датасет Wine из sklearn как пример
from sklearn.datasets import load_wine
data = load_wine()
X = data.data
y = data.target
feature_names = data.feature_names
target_names = data.target_names

print(f"Датасет: Wine Dataset")
print(f"Количество образцов: {X.shape[0]}")
print(f"Количество признаков: {X.shape[1]}")
print(f"Количество классов: {len(np.unique(y))}")
print(f"Названия классов: {target_names}")
print(f"Названия признаков: {feature_names}")

# Создаем DataFrame для удобства
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

print("\nПервые 5 строк датасета:")
print(df.head())

print("\nСтатистика датасета:")
print(df.describe())

print("\nРаспределение классов:")
print(df['target'].value_counts())

# Проверка на пропущенные значения
print(f"\nПропущенные значения: {df.isnull().sum().sum()}")

# Визуализация распределения признаков
plt.figure(figsize=(15, 10))
for i, feature in enumerate(feature_names[:6]):
    plt.subplot(2, 3, i+1)
    for target_class in np.unique(y):
        plt.hist(X[y == target_class, i], alpha=0.5, label=target_names[target_class])
    plt.xlabel(feature)
    plt.ylabel('Частота')
    plt.legend()
plt.suptitle('Распределение признаков по классам')
plt.tight_layout()
plt.savefig('feature_distribution.png', dpi=300)
plt.show()

# Проверка на выбросы с помощью z-score
z_scores = np.abs(stats.zscore(X))
outliers = np.where(z_scores > 3)
print(f"\nКоличество выбросов (z-score > 3): {len(outliers[0])}")

# ============================================================================
# 2. РАЗБИЕНИЕ ВЫБОРКИ
# ============================================================================

print("\n" + "=" * 80)
print("2. РАЗБИЕНИЕ ВЫБОРКИ")
print("=" * 80)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Размер обучающей выборки: {X_train.shape}")
print(f"Размер тестовой выборки: {X_test.shape}")
print(f"Распределение классов в обучающей выборке:")
print(pd.Series(y_train).value_counts())
print(f"Распределение классов в тестовой выборке:")
print(pd.Series(y_test).value_counts())

# Масштабирование признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ============================================================================
# 3. МЕТОДЫ КЛАССИФИКАЦИИ И НАСТРОЙКА ГИПЕРПАРАМЕТРОВ
# ============================================================================

print("\n" + "=" * 80)
print("3. МЕТОДЫ КЛАССИФИКАЦИИ И НАСТРОЙКА ГИПЕРПАРАМЕТРОВ")
print("=" * 80)

# Создаем словарь для хранения результатов
results = {}

def evaluate_model(model, model_name, X_train, X_test, y_train, y_test):
    """Оценка модели и сохранение метрик"""
    # Обучение модели
    model.fit(X_train, y_train)
    
    # Предсказания
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None
    
    # Вычисление метрик
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted'),
        'recall': recall_score(y_test, y_pred, average='weighted'),
        'f1': f1_score(y_test, y_pred, average='weighted'),
    }
    
    # ROC-AUC (только для бинарной классификации или если есть predict_proba)
    if y_pred_proba is not None and len(np.unique(y_test)) == 2:
        metrics['roc_auc'] = roc_auc_score(y_test, y_pred_proba[:, 1])
    elif y_pred_proba is not None:
        # Для многоклассовой классификации
        metrics['roc_auc'] = roc_auc_score(y_test, y_pred_proba, multi_class='ovr')
    else:
        metrics['roc_auc'] = None
    
    # Сохранение результатов
    results[model_name] = {
        'model': model,
        'metrics': metrics,
        'predictions': y_pred
    }
    
    print(f"\n{model_name}:")
    print(f"  Accuracy: {metrics['accuracy']:.4f}")
    print(f"  Precision: {metrics['precision']:.4f}")
    print(f"  Recall: {metrics['recall']:.4f}")
    print(f"  F1-Score: {metrics['f1']:.4f}")
    if metrics['roc_auc'] is not None:
        print(f"  ROC-AUC: {metrics['roc_auc']:.4f}")
    
    return metrics

# ----------------------------------------------------------------------------
# 3.1. НАИВНЫЙ БАЙЕСОВСКИЙ КЛАССИФИКАТОР
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("3.1. НАИВНЫЙ БАЙЕСОВСКИЙ КЛАССИФИКАТОР")
print("-" * 80)

# Для GaussianNB нужны масштабированные данные
nb_gaussian = GaussianNB()
evaluate_model(nb_gaussian, "GaussianNB", X_train_scaled, X_test_scaled, y_train, y_test)

# Для MultinomialNB нужны положительные значения
X_train_nonneg = X_train_scaled - X_train_scaled.min() + 0.1
X_test_nonneg = X_test_scaled - X_test_scaled.min() + 0.1
nb_multinomial = MultinomialNB()
evaluate_model(nb_multinomial, "MultinomialNB", X_train_nonneg, X_test_nonneg, y_train, y_test)

nb_complement = ComplementNB()
evaluate_model(nb_complement, "ComplementNB", X_train_nonneg, X_test_nonneg, y_train, y_test)

nb_bernoulli = BernoulliNB()
evaluate_model(nb_bernoulli, "BernoulliNB", X_train_scaled, X_test_scaled, y_train, y_test)

# Настройка гиперпараметров для GaussianNB
print("\nНастройка гиперпараметров для GaussianNB:")
param_grid_nb = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}
grid_nb = GridSearchCV(GaussianNB(), param_grid_nb, cv=5, scoring='accuracy')
grid_nb.fit(X_train_scaled, y_train)
print(f"Лучшие параметры: {grid_nb.best_params_}")
print(f"Лучшая точность на кросс-валидации: {grid_nb.best_score_:.4f}")

# Обучение с лучшими параметрами
nb_tuned = GaussianNB(**grid_nb.best_params_)
evaluate_model(nb_tuned, "GaussianNB (Tuned)", X_train_scaled, X_test_scaled, y_train, y_test)

# ----------------------------------------------------------------------------
# 3.2. ДЕРЕВЬЯ РЕШЕНИЙ
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("3.2. ДЕРЕВЬЯ РЕШЕНИЙ")
print("-" * 80)

# Базовая модель
dt_base = DecisionTreeClassifier(random_state=42)
evaluate_model(dt_base, "Decision Tree (Base)", X_train, X_test, y_train, y_test)

# Настройка гиперпараметров
print("\nНастройка гиперпараметров для Decision Tree:")
param_grid_dt = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), 
                      param_grid_dt, cv=5, scoring='accuracy')
grid_dt.fit(X_train, y_train)
print(f"Лучшие параметры: {grid_dt.best_params_}")
print(f"Лучшая точность на кросс-валидации: {grid_dt.best_score_:.4f}")

# Обучение с лучшими параметрами
dt_tuned = DecisionTreeClassifier(**grid_dt.best_params_, random_state=42)
evaluate_model(dt_tuned, "Decision Tree (Tuned)", X_train, X_test, y_train, y_test)

# Визуализация дерева (упрощенная)
plt.figure(figsize=(20, 10))
from sklearn.tree import plot_tree
plot_tree(dt_tuned, feature_names=feature_names, 
          class_names=target_names, filled=True, rounded=True, max_depth=3)
plt.title("Дерево решений (первые 3 уровня)")
plt.tight_layout()
plt.savefig('decision_tree.png', dpi=300)
plt.show()

# ----------------------------------------------------------------------------
# 3.3. ЛИНЕЙНЫЙ ДИСКРИМИНАНТНЫЙ АНАЛИЗ
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("3.3. ЛИНЕЙНЫЙ ДИСКРИМИНАНТНЫЙ АНАЛИЗ")
print("-" * 80)

# Базовая модель
lda_base = LinearDiscriminantAnalysis()
evaluate_model(lda_base, "LDA (Base)", X_train_scaled, X_test_scaled, y_train, y_test)

# Настройка гиперпараметров
print("\nНастройка гиперпараметров для LDA:")
param_grid_lda = {
    'solver': ['svd', 'lsqr', 'eigen'],
    'shrinkage': [None, 'auto', 0.1, 0.5, 0.9]
}
grid_lda = GridSearchCV(LinearDiscriminantAnalysis(), 
                       param_grid_lda, cv=5, scoring='accuracy')
grid_lda.fit(X_train_scaled, y_train)
print(f"Лучшие параметры: {grid_lda.best_params_}")
print(f"Лучшая точность на кросс-валидации: {grid_lda.best_score_:.4f}")

# Обучение с лучшими параметрами
lda_tuned = LinearDiscriminantAnalysis(**grid_lda.best_params_)
evaluate_model(lda_tuned, "LDA (Tuned)", X_train_scaled, X_test_scaled, y_train, y_test)

# ----------------------------------------------------------------------------
# 3.4. МЕТОД ОПОРНЫХ ВЕКТОРОВ
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("3.4. МЕТОД ОПОРНЫХ ВЕКТОРОВ")
print("-" * 80)

# Базовая модель
svm_base = SVC(probability=True, random_state=42)
evaluate_model(svm_base, "SVM (Base)", X_train_scaled, X_test_scaled, y_train, y_test)

# Настройка гиперпараметров
print("\nНастройка гиперпараметров для SVM:")
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto', 0.1, 1],
    'degree': [2, 3, 4]  # только для poly
}
grid_svm = GridSearchCV(SVC(probability=True, random_state=42), 
                       param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train_scaled, y_train)
print(f"Лучшие параметры: {grid_svm.best_params_}")
print(f"Лучшая точность на кросс-валидации: {grid_svm.best_score_:.4f}")

# Обучение с лучшими параметрами
svm_tuned = SVC(**grid_svm.best_params_, probability=True, random_state=42)
evaluate_model(svm_tuned, "SVM (Tuned)", X_train_scaled, X_test_scaled, y_train, y_test)

# ----------------------------------------------------------------------------
# 3.5. МЕТОД БЛИЖАЙШИХ СОСЕДЕЙ
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("3.5. МЕТОД БЛИЖАЙШИХ СОСЕДЕЙ")
print("-" * 80)

# Базовая модель
knn_base = KNeighborsClassifier()
evaluate_model(knn_base, "KNN (Base)", X_train_scaled, X_test_scaled, y_train, y_test)

# Настройка гиперпараметров
print("\nНастройка гиперпараметров для KNN:")
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2]  # 1: манхэттенское расстояние, 2: евклидово
}
grid_knn = GridSearchCV(KNeighborsClassifier(), 
                       param_grid_knn, cv=5, scoring='accuracy')
grid_knn.fit(X_train_scaled, y_train)
print(f"Лучшие параметры: {grid_knn.best_params_}")
print(f"Лучшая точность на кросс-валидации: {grid_knn.best_score_:.4f}")

# Обучение с лучшими параметрами
knn_tuned = KNeighborsClassifier(**grid_knn.best_params_)
evaluate_model(knn_tuned, "KNN (Tuned)", X_train_scaled, X_test_scaled, y_train, y_test)

# Исследование влияния числа соседей
neighbors_range = range(1, 21)
train_scores = []
test_scores = []

for n in neighbors_range:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train_scaled, y_train)
    train_scores.append(knn.score(X_train_scaled, y_train))
    test_scores.append(knn.score(X_test_scaled, y_test))

plt.figure(figsize=(10, 6))
plt.plot(neighbors_range, train_scores, label='Обучающая выборка')
plt.plot(neighbors_range, test_scores, label='Тестовая выборка')
plt.xlabel('Число соседей')
plt.ylabel('Точность')
plt.title('Влияние числа соседей на точность KNN')
plt.legend()
plt.grid(True)
plt.savefig('knn_neighbors_impact.png', dpi=300)
plt.show()

# ============================================================================
# 4. НЕЙРОННАЯ СЕТЬ НА TENSORFLOW
# ============================================================================

print("\n" + "=" * 80)
print("4. НЕЙРОННАЯ СЕТЬ НА TENSORFLOW")
print("=" * 80)

# Подготовка данных для нейронной сети
# One-hot encoding для целевой переменной
y_train_nn = tf.keras.utils.to_categorical(y_train, num_classes=3)
y_test_nn = tf.keras.utils.to_categorical(y_test, num_classes=3)

# ----------------------------------------------------------------------------
# 4.1. БАЗОВАЯ НЕЙРОННАЯ СЕТЬ
# ----------------------------------------------------------------------------

def create_baseline_nn():
    """Создание базовой нейронной сети"""
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        layers.Dropout(0.3),
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(16, activation='relu'),
        layers.Dense(3, activation='softmax')
    ])
    
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Настройка TensorBoard
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

# Создание и обучение базовой модели
print("\nОбучение базовой нейронной сети...")
baseline_nn = create_baseline_nn()

# Показываем архитектуру
baseline_nn.summary()

history_baseline = baseline_nn.fit(
    X_train_scaled, y_train_nn,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    verbose=1,
    callbacks=[
        tensorboard_callback, 
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    ]
)

# Оценка базовой модели
test_loss, test_accuracy = baseline_nn.evaluate(X_test_scaled, y_test_nn, verbose=0)
print(f"\nБазовая нейронная сеть - Точность на тесте: {test_accuracy:.4f}")

# Визуализация процесса обучения
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# График точности
axes[0].plot(history_baseline.history['accuracy'], label='Обучающая точность')
axes[0].plot(history_baseline.history['val_accuracy'], label='Валидационная точность')
axes[0].set_xlabel('Эпоха')
axes[0].set_ylabel('Точность')
axes[0].set_title('Точность обучения')
axes[0].legend()
axes[0].grid(True)

# График потерь
axes[1].plot(history_baseline.history['loss'], label='Обучающие потери')
axes[1].plot(history_baseline.history['val_loss'], label='Валидационные потери')
axes[1].set_xlabel('Эпоха')
axes[1].set_ylabel('Потери')
axes[1].set_title('Потери обучения')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('nn_training_history.png', dpi=300)
plt.show()

# ----------------------------------------------------------------------------
# 4.2. РУЧНАЯ КРОСС-ВАЛИДАЦИЯ ДЛЯ НЕЙРОННОЙ СЕТИ
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("4.2. 5-КРАТНАЯ КРОСС-ВАЛИДАЦИЯ С РУЧНЫМ ПОДБОРОМ ПАРАМЕТРОВ")
print("-" * 80)

# Определяем различные архитектуры для тестирования
architectures = [
    {'hidden_layers': 1, 'units': 32, 'dropout': 0.2},
    {'hidden_layers': 2, 'units': 64, 'dropout': 0.3},
    {'hidden_layers': 3, 'units': 128, 'dropout': 0.4}
]

learning_rates = [0.001, 0.01, 0.1]

cv_scores = []
best_score = 0
best_params = None
best_model = None

# Стратифицированная кросс-валидация
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("Начинаем кросс-валидацию с различными параметрами...")

for lr in learning_rates:
    for arch in architectures:
        print(f"\nТестирование: LR={lr}, Архитектура={arch}")
        
        fold_scores = []
        
        for fold_idx, (train_idx, val_idx) in enumerate(skf.split(X_train_scaled, y_train)):
            # Разделение данных
            X_train_fold = X_train_scaled[train_idx]
            X_val_fold = X_train_scaled[val_idx]
            y_train_fold = tf.keras.utils.to_categorical(y_train[train_idx], num_classes=3)
            y_val_fold = tf.keras.utils.to_categorical(y_train[val_idx], num_classes=3)
            
            # Создание модели
            model = models.Sequential()
            model.add(layers.Dense(arch['units'], activation='relu', 
                                   input_shape=(X_train_scaled.shape[1],)))
            model.add(layers.Dropout(arch['dropout']))
            
            for i in range(arch['hidden_layers'] - 1):
                units = arch['units'] // (2 ** (i + 1))
                model.add(layers.Dense(units, activation='relu'))
                model.add(layers.Dropout(arch['dropout']))
            
            model.add(layers.Dense(3, activation='softmax'))
            
            # Компиляция
            optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
            model.compile(
                optimizer=optimizer,
                loss='categorical_crossentropy',
                metrics=['accuracy']
            )
            
            # Обучение
            history = model.fit(
                X_train_fold, y_train_fold,
                epochs=50,
                batch_size=16,
                validation_data=(X_val_fold, y_val_fold),
                verbose=0,
                callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)]
            )
            
            # Оценка
            val_loss, val_accuracy = model.evaluate(X_val_fold, y_val_fold, verbose=0)
            fold_scores.append(val_accuracy)
            
            print(f"  Fold {fold_idx + 1}: accuracy = {val_accuracy:.4f}")
        
        # Средняя точность по фолдам
        mean_score = np.mean(fold_scores)
        cv_scores.append({
            'learning_rate': lr,
            'architecture': arch,
            'mean_accuracy': mean_score,
            'fold_scores': fold_scores
        })
        
        print(f"  Средняя точность: {mean_score:.4f}")
        
        # Сохраняем лучшую модель
        if mean_score > best_score:
            best_score = mean_score
            best_params = {'learning_rate': lr, 'architecture': arch}
            best_model = model

# Вывод результатов кросс-валидации
print("\n" + "=" * 80)
print("РЕЗУЛЬТАТЫ КРОСС-ВАЛИДАЦИИ")
print("=" * 80)

for i, result in enumerate(cv_scores):
    print(f"\nКонфигурация {i+1}:")
    print(f"  Learning rate: {result['learning_rate']}")
    print(f"  Архитектура: {result['architecture']}")
    print(f"  Точность по фолдам: {[f'{score:.4f}' for score in result['fold_scores']]}")
    print(f"  Средняя точность: {result['mean_accuracy']:.4f}")

print(f"\nЛучшая конфигурация:")
print(f"  Learning rate: {best_params['learning_rate']}")
print(f"  Архитектура: {best_params['architecture']}")
print(f"  Средняя точность на кросс-валидации: {best_score:.4f}")

# ----------------------------------------------------------------------------
# 4.3. ОБУЧЕНИЕ И ОЦЕНКА ЛУЧШЕЙ МОДЕЛИ
# ----------------------------------------------------------------------------

print("\n" + "-" * 80)
print("4.3. ОБУЧЕНИЕ И ОЦЕНКА ЛУЧШЕЙ МОДЕЛИ НА ВСЕХ ДАННЫХ")
print("-" * 80)

# Создаем и обучаем лучшую модель на всех обучающих данных
best_arch = best_params['architecture']
best_lr = best_params['learning_rate']

final_model = models.Sequential()
final_model.add(layers.Dense(best_arch['units'], activation='relu', 
                            input_shape=(X_train_scaled.shape[1],)))
final_model.add(layers.Dropout(best_arch['dropout']))

for i in range(best_arch['hidden_layers'] - 1):
    units = best_arch['units'] // (2 ** (i + 1))
    final_model.add(layers.Dense(units, activation='relu'))
    final_model.add(layers.Dropout(best_arch['dropout']))

final_model.add(layers.Dense(3, activation='softmax'))

optimizer = tf.keras.optimizers.Adam(learning_rate=best_lr)
final_model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Настройка TensorBoard для финальной модели
log_dir_final = f"logs/final_model/{datetime.datetime.now().strftime('%Y%m%d-%H%M%S')}"
tensorboard_final = TensorBoard(log_dir=log_dir_final)

# Обучение финальной модели
history_final = final_model.fit(
    X_train_scaled, y_train_nn,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    verbose=1,
    callbacks=[
        tensorboard_final,
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    ]
)

# Оценка финальной модели
test_loss_final, test_accuracy_final = final_model.evaluate(X_test_scaled, y_test_nn, verbose=0)
print(f"\nФинальная нейронная сеть - Точность на тесте: {test_accuracy_final:.4f}")
print(f"Улучшение по сравнению с базовой моделью: {test_accuracy_final - test_accuracy:.4f}")

# Визуализация обучения финальной модели
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# График точности
axes[0].plot(history_final.history['accuracy'], label='Обучающая точность')
axes[0].plot(history_final.history['val_accuracy'], label='Валидационная точность')
axes[0].set_xlabel('Эпоха')
axes[0].set_ylabel('Точность')
axes[0].set_title('Точность обучения (финальная модель)')
axes[0].legend()
axes[0].grid(True)

# График потерь
axes[1].plot(history_final.history['loss'], label='Обучающие потери')
axes[1].plot(history_final.history['val_loss'], label='Валидационные потери')
axes[1].set_xlabel('Эпоха')
axes[1].set_ylabel('Потери')
axes[1].set_title('Потери обучения (финальная модель)')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('nn_final_training_history.png', dpi=300)
plt.show()

# Сохранение результатов нейронной сети
y_pred_nn = final_model.predict(X_test_scaled)
y_pred_nn_classes = np.argmax(y_pred_nn, axis=1)

results['Neural Network (Baseline)'] = {
    'model': baseline_nn,
    'metrics': {
        'accuracy': test_accuracy,
        'precision': precision_score(y_test, np.argmax(baseline_nn.predict(X_test_scaled), axis=1), average='weighted'),
        'recall': recall_score(y_test, np.argmax(baseline_nn.predict(X_test_scaled), axis=1), average='weighted'),
        'f1': f1_score(y_test, np.argmax(baseline_nn.predict(X_test_scaled), axis=1), average='weighted'),
        'roc_auc': roc_auc_score(y_test_nn, baseline_nn.predict(X_test_scaled), multi_class='ovr')
    }
}

results['Neural Network (Tuned)'] = {
    'model': final_model,
    'metrics': {
        'accuracy': test_accuracy_final,
        'precision': precision_score(y_test, y_pred_nn_classes, average='weighted'),
        'recall': recall_score(y_test, y_pred_nn_classes, average='weighted'),
        'f1': f1_score(y_test, y_pred_nn_classes, average='weighted'),
        'roc_auc': roc_auc_score(y_test_nn, y_pred_nn, multi_class='ovr')
    }
}

# ============================================================================
# 5. СРАВНИТЕЛЬНЫЙ АНАЛИЗ
# ============================================================================

print("\n" + "=" * 80)
print("5. СРАВНИТЕЛЬНЫЙ АНАЛИЗ")
print("=" * 80)

# Создание сравнительной таблицы
comparison_df = pd.DataFrame()

for model_name, model_data in results.items():
    metrics = model_data['metrics']
    comparison_df[model_name] = pd.Series({
        'Accuracy': metrics['accuracy'],
        'Precision': metrics['precision'],
        'Recall': metrics['recall'],
        'F1-Score': metrics['f1'],
        'ROC-AUC': metrics['roc_auc'] if metrics['roc_auc'] is not None else np.nan
    })

# Транспонируем для удобства
comparison_df = comparison_df.T

print("\nСРАВНИТЕЛЬНАЯ ТАБЛИЦА МЕТРИК:")
print("-" * 80)
print(comparison_df.round(4))

# Визуализация сравнения
plt.figure(figsize=(15, 8))
comparison_df[['Accuracy', 'Precision', 'Recall', 'F1-Score']].plot(kind='bar', figsize=(15, 8))
plt.title('Сравнение метрик классификаторов')
plt.ylabel('Значение метрики')
plt.xlabel('Модель')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('models_comparison.png', dpi=300)
plt.show()

# Тепловая карта метрик
plt.figure(figsize=(12, 8))
sns.heatmap(comparison_df, annot=True, fmt='.4f', cmap='YlOrRd', center=0.9)
plt.title('Тепловая карта метрик качества моделей')
plt.tight_layout()
plt.savefig('metrics_heatmap.png', dpi=300)
plt.show()

# Определение лучшей модели по accuracy
best_model_name = comparison_df['Accuracy'].idxmax()
best_model_accuracy = comparison_df['Accuracy'].max()

print("\n" + "=" * 80)
print("ВЫВОДЫ")
print("=" * 80)
print(f"Лучшая модель: {best_model_name}")
print(f"Точность лучшей модели: {best_model_accuracy:.4f}")
print("\nРейтинг моделей по точности:")
print("-" * 40)
for i, (model, acc) in enumerate(comparison_df['Accuracy'].sort_values(ascending=False).items(), 1):
    print(f"{i}. {model}: {acc:.4f}")

# ============================================================================
# 6. ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ
# ============================================================================

print("\n" + "=" * 80)
print("6. ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ")
print("=" * 80)

# Матрицы ошибок для лучших моделей каждого типа
best_models = {
    'GaussianNB': results['GaussianNB (Tuned)']['model'],
    'Decision Tree': results['Decision Tree (Tuned)']['model'],
    'LDA': results['LDA (Tuned)']['model'],
    'SVM': results['SVM (Tuned)']['model'],
    'KNN': results['KNN (Tuned)']['model'],
    'Neural Network': results['Neural Network (Tuned)']['model']
}

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for idx, (name, model) in enumerate(best_models.items()):
    if hasattr(model, 'predict'):
        if name in ['GaussianNB', 'LDA']:
            y_pred = model.predict(X_test_scaled)
        elif name == 'Neural Network':
            y_pred = np.argmax(model.predict(X_test_scaled), axis=1)
        else:
            y_pred = model.predict(X_test)
        
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                   xticklabels=target_names, yticklabels=target_names)
        axes[idx].set_title(f'Матрица ошибок: {name}')
        axes[idx].set_xlabel('Предсказанный класс')
        axes[idx].set_ylabel('Истинный класс')

plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=300)
plt.show()

# Кривые ROC для моделей с вероятностями
plt.figure(figsize=(12, 10))
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink']

for idx, (name, model_data) in enumerate(results.items()):
    model = model_data['model']
    
    if hasattr(model, 'predict_proba'):
        # Для классических моделей
        y_proba = model.predict_proba(X_test_scaled)
    elif hasattr(model, 'predict'):
        # Для нейронных сетей
        y_proba = model.predict(X_test_scaled)
    else:
        continue
    
    # Для многоклассовой классификации, строим ROC для каждого класса
    if len(np.unique(y_test)) > 2:
        # Используем one-vs-rest подход
        y_test_binarized = tf.keras.utils.to_categorical(y_test, num_classes=3)
        roc_auc_scores = []
        for i in range(3):
            fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_proba[:, i])
            roc_auc = roc_auc_score(y_test_binarized[:, i], y_proba[:, i])
            roc_auc_scores.append(roc_auc)
            if i == 0:  # строим только для первого класса для читаемости
                plt.plot(fpr, tpr, color=colors[idx % len(colors)], alpha=0.7,
                        label=f'{name} (Avg AUC={np.mean(roc_auc_scores):.3f})')
    else:
        # Для бинарной классификации
        fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1])
        roc_auc = roc_auc_score(y_test, y_proba[:, 1])
        plt.plot(fpr, tpr, color=colors[idx % len(colors)], 
                label=f'{name} AUC={roc_auc:.3f}')

plt.plot([0, 1], [0, 1], 'k--', label='Случайный классификатор (AUC=0.5)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Кривые ROC для различных моделей')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig('roc_curves.png', dpi=300)
plt.show()

# ============================================================================
# 7. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
# ============================================================================

print("\n" + "=" * 80)
print("7. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ")
print("=" * 80)

# Сохранение сравнения в CSV
comparison_df.to_csv('models_comparison.csv')
print("Результаты сохранены в файл 'models_comparison.csv'")

# Сохранение лучшей модели
final_model.save('best_neural_network.h5')
print("Лучшая нейронная сеть сохранена в файл 'best_neural_network.h5'")

# Сохранение всех моделей
import joblib
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(best_models, 'best_models.pkl')
print("Скалер и лучшие модели сохранены")

# Сохранение графиков
print("\nГрафики сохранены в файлы:")
print("- feature_distribution.png")
print("- decision_tree.png")
print("- knn_neighbors_impact.png")
print("- nn_training_history.png")
print("- nn_final_training_history.png")
print("- models_comparison.png")
print("- metrics_heatmap.png")
print("- confusion_matrices.png")
print("- roc_curves.png")

print("\n" + "=" * 80)
print("ЛАБОРАТОРНАЯ РАБОТА ВЫПОЛНЕНА")
print("=" * 80)

# Инструкция для TensorBoard
print("\n" + "=" * 80)
print("ИНСТРУКЦИЯ ДЛЯ TENSORBOARD")
print("=" * 80)
print("Для просмотра графиков обучения в TensorBoard выполните в терминале:")
print("tensorboard --logdir logs/")
print("Затем откройте браузер и перейдите по адресу: http://localhost:6006")

# Генерация отчета
print("\n" + "=" * 80)
print("КРАТКИЙ ОТЧЕТ")
print("=" * 80)
print(f"Датасет: Wine Dataset ({X.shape[0]} образцов, {X.shape[1]} признаков)")
print(f"Количество классов: {len(np.unique(y))}")
print(f"Лучшая модель: {best_model_name}")
print(f"Точность лучшей модели: {best_model_accuracy:.4f}")
print(f"Улучшение настройки нейронной сети: {test_accuracy_final - test_accuracy:.4f}")
```
